```python
# basic example of SELECT CASE (WHEN THEN ELSE statement)
SELECT Name, Surname, Salary
CASE
    WHEN Salary >= 0 AND Salary <= 1500 THEN 0.1*Salary
    WHEN Salary >= 1500 AND Salary <= 2500 THEN 0.15*Salary
    WHEN Salary >= 2500 THEN 0.2*Salary
    ELSE 0
    END AS Tax
FROM Owners
ORDER BY Surname ASC
```
