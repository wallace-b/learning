```python
# Load dependencies
from string import ascii_letters, digits, punctuation
from itertools import product
```

```python
character_options = ascii_letters + digits + punctuation # = 94 chars, where 26 lower + 26 UPPER + 10 digits + 32 punctuations  
print (f"There are {len(character_options)} choices for a single character.")
```

```python
# Example #1 - 4 digit, 8 letter, 8 character passcodes 

# Number of combinations with repetition/replacement
print (f"For a 4 digit passcode, there are {len(digits)**4} possible combinations with repetition/replacement.")
print (f"For a 8 letter passcode, there are {len(ascii_letters)**8} possible combinations with repetition/replacement.")
print (f"For a 8 character passcode, there are {len(character_options)**8} possible combinations with repetition/replacement.")
```

```python
# Example #2 - Brute force search

loop_limit = 200 # for brevity
counter = 0

for guess_passcode in product(character_options, repeat=8):
    print(*guess_passcode)
    counter += 1
    if counter == 200: break
```

### Protections against brute force adversaries
In general, there is generally some % chance of security breach (even if infinitesimally small). We raise the bar of security & raise the cost or time required or compute of breaching.

We can increase the cost, time and compute for a breach with a number of techniques:
* Limited attempts, lock-out periods and locking of devices until a phone call
* 2FA - using an email, smartphone app or FOB key

### Protections against leaking information
Your computer is always leaking information, it is just a matter of controlling that sensitive information. 

There are number of techniques to control this sensitive information leaking:
* In a limited way, incognito mode does not track history and cookies (the 'memory' (passwords, search preferences, and so on) on websites and web-browsing). However, it is still possible for your organization or a website to identify you via private IP (local IP) or public IP (external IP).
* https -> http's' sites, the 's' means 'secure' and the conversation between the user and the server is encrypted i.e. end-to-end encryption.
